## Part 1: Data Cleaning

### Convert Total Stats to Per Game Stats

```python
player_stats['MIN'] = round(player_stats['MIN'] / player_stats['GP'], 1)
player_stats['PTS'] = round(player_stats['PTS'] / player_stats['GP'], 1)
player_stats['AST'] = round(player_stats['AST'] / player_stats['GP'], 1)
player_stats['REB'] = round(player_stats['REB'] / player_stats['GP'], 1)
player_stats['STL'] = round(player_stats['STL'] / player_stats['GP'], 1)
player_stats['BLK'] = round(player_stats['BLK'] / player_stats['GP'], 1)
```

### Map Team Abbreviations to Team Names

```python
map_team_names(player_stats, team_stats)
```

```python
# view result
player_stats.head()
```

### Merge Dataframes

```python
# merge player and team stats
player_team_stats = player_stats.merge(team_stats, on=['Season', 'TEAM_NAME'], how='left')
```

```python
print(f"player_stats: {player_stats.shape}")
print(f"player_team_stats: {player_team_stats.shape}")
```

```python
# rename player name col in mvp shares to match player_team stats
mvp_shares = mvp_shares.rename(columns={'Player Name': 'PLAYER_NAME'})
```

```python
# find any name mismatches between bbr and nba_api
mvp_names = set(mvp_shares['PLAYER_NAME'].unique())
player_names = set(player_team_stats['PLAYER_NAME'].unique())

unmatched = mvp_names - player_names
print(unmatched)

for name in ['Dragic', 'Ginobili', 'Stojakovic', 'Butler']:
    matches = player_team_stats[player_team_stats['PLAYER_NAME'].str.contains(name)]
    print(matches['PLAYER_NAME'].unique())

# normalize names
player_name_normalized = {
    'Goran Dragić': 'Goran Dragic',
    'Manu Ginóbili': 'Manu Ginobili',
    'Peja Stojaković': 'Peja Stojakovic',
    'Jimmy Butler': 'Jimmy Butler III'
}

# apply mapping
mvp_shares['PLAYER_NAME'] = mvp_shares['PLAYER_NAME'].replace(player_name_normalized)
```

```python
# confirm mapping was successful
unmatched = set(mvp_shares['PLAYER_NAME'].unique()) - set(player_team_stats['PLAYER_NAME'].unique())
print(unmatched)
```

```python
# merge mvp_shares with player_team_stats
merged_df = player_team_stats.merge(mvp_shares, on=['Season', 'PLAYER_NAME'], how='left')

print(merged_df.head())
```

```python
merged_df.columns
```

### Clean Merged Dataframe

```python
# drop irrelevant cols
cols_to_drop = [
    'Unnamed: 0_x',   # index artifact
    'Unnamed: 0_y',   # index artifact
    'Unnamed: 0',     # index artifact
    'W_x',            # redundant with W_y
    'L_x',            # redundant with L_y
    'W_PCT_RANK',      # redundant with W_RANK
    'Rank',           # from mvp_shares, not a feature
    'Team',           # redundant with TEAM_NAME
    'TEAM_ABBREVIATION', # redundant with TEAM_NAME
    'First Place Votes',      # leakage - derived from MVP share
    'Total Voting Points',    # leakage - derived from MVP share
    'Voting_Pts Max',         # constant, not useful
]
merged_df = merged_df.drop(columns=cols_to_drop)

# rename cols for consistency and readability
merged_df = merged_df.rename(columns={'W_y': 'TEAM_W', 'L_y': 'TEAM_L', 'MVP Share': 'MVP_SHARE', 'Season': 'SEASON'})

# fill all NaNs in mvp share with 0s - these players recieved no mvp votes
merged_df['MVP_SHARE'] = merged_df['MVP_SHARE'].fillna(0)
```

```python
# observe resulting dataframe
print(merged_df.shape)
print(merged_df.columns.tolist())
print(f"Non-zero MVP shares: {(merged_df['MVP_SHARE'] > 0).sum()}")
print(merged_df.isnull().sum())
```

```python
# save merged dataframe to a csv file
merged_df.to_csv('../data/processed/merged_df.csv')
```
